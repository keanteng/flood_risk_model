```python
import pandas as pd
import geopandas as gpd
```

```python
geocoded = pd.read_csv('data/random_address_geocoded.csv')
flood_points = pd.read_csv('data/all_states_all_years_geocoded.csv')
```

```python
geocoded_data = gpd.GeoDataFrame(geocoded, geometry = gpd.points_from_xy(geocoded.Longitude, geocoded.Latitude), crs = "EPSG:4326").to_crs('EPSG:3857')
flood_points_data = gpd.GeoDataFrame(flood_points, geometry = gpd.points_from_xy(flood_points.Longitude, flood_points.Latitude), crs = "EPSG:4326").to_crs('EPSG:3857')
```

```python
geocoded_data.head()
```

```python
flood_points_data.head()
```

```python
# Select a random location
random_location = geocoded_data.iloc[360]

# Measure distance from release to each station
distances = flood_points_data.geometry.distance(random_location.geometry)
distances
```

```python
print("Mean distance: {}".format(distances.mean()))
print("Nearest flood point: {}".format(distances.min()))

# find that nearest flood point
print(flood_points_data.iloc[distances.idxmin()][["Name", "Latitude", "Longitude"]])
```

```python
# write a function to compute for the minimum distances from any flood point
random_location = geocoded_data.iloc[360]
distances = flood_points_data.geometry.distance(random_location.geometry)
distances.min()

for i in range(0, len(geocoded_data)):
    distances = flood_points_data.geometry.distance(geocoded_data.iloc[i].geometry)
    geocoded_data.loc[i, 'distance'] = distances.min()
```

```python
geocoded_data.head()
```

```python
data = pd.DataFrame()

data['distance'] = geocoded_data['distance']
data['city'] = geocoded_data['City']
data['state'] = geocoded_data['State']
data['zip'] = geocoded_data['Zip']
data.head(2)
```

```python
data['flood_risk'] = data['distance'].apply(lambda x: 0 if x > 500 else 1)
data.head()
```

```python
data.to_csv('data/predictor.csv', index=False)
```
