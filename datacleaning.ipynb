```python
import pandas as pd
```

```python
data = pd.read_excel('random_address7.xlsx')
data.head(2)
```

```python
data['Address'] = data['Address'].str.split(" ;", expand=False)
data.head()
```

```python
# cleaning the address line
new1 = [None]*len(data)
new2 = [None]*len(data)
new3 = [None]*len(data)
new4 = [None]*len(data)
new5 = [None]*len(data)
new6 = [None]*len(data)
new7 = [None]*len(data)

for i in range(0, len(data)):
    new1[i] = data['Address'][i][0]
    new1[i] = new1[i].replace('Street:  ', '')

for i in range(0, len(data)):
    new2[i] = data['Address'][i][1]
    new2[i] = new2[i].replace('City:   ', '')

for i in range(0, len(data)):
    new3[i] = data['Address'][i][2]
    new3[i] = new3[i].replace('State/province/area:    ', '')
    
for i in range(0, len(data)):
    new4[i] = data['Address'][i][3]
    new4[i] = new4[i].replace('Phone number  ', '')
    
for i in range(0, len(data)):
    new5[i] = data['Address'][i][4]
    new5[i] = new5[i].replace('Zip code  ', '')
    
for i in range(0, len(data)):
    new6[i] = data['Address'][i][5]
    new6[i] = new6[i].replace('Country calling code  ', '')
    
for i in range(0, len(data)):
    new7[i] = data['Address'][i][6]
    new7[i] = new7[i].replace('Country  ', '')

data['Place'] = new1
data['City'] = new2
data['State'] = new3
data['Phone'] = new4
data['Zip'] = new5
data['CountryCode'] = new6
data['Country'] = new7

data.head()
```

```python
data['Location'] = data['Place'] + ', ' + data['City'] + ', ' + data['State']
data.head(2)
```

```python
data.to_csv('random_address_cleaned7.csv', index=False)
```
